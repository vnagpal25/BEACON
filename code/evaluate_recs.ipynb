```python
from meal_recommender import MealRecommender
```

```python
random_goodness = {}
bandit_goodness = {}

for i in range(1,26):
    user_reco_evaluator = MealRecommender(f'user_{i}.json')
    
    random_rec = f'random_recs/recommendation_user_{i}.json'
    bandit_rec = f'bandit_recs/recommendation_user_{i}.json'
    
    user_reco_evaluator.SetRecommendation(bandit_rec)
    user_reco_evaluator.EvaluateRecs()
    bandit_goodness[f"{user_reco_evaluator.recommendation['goodness']}_{i}"] = user_reco_evaluator.score_breakdown
    print('------------')
    
    user_reco_evaluator.SetRecommendation(random_rec)
    user_reco_evaluator.EvaluateRecs()
    random_goodness[f"{user_reco_evaluator.recommendation['goodness']}_{i}"] = user_reco_evaluator.score_breakdown
    print('------------')
```

```python
import json
```

```python
with open('../recommendations/eval/bandit_res.json', 'w') as file:
    json.dump(bandit_goodness, file)
```

```python
with open('../recommendations/eval/random_res.json', 'w') as file:
    json.dump(random_goodness, file)
```
